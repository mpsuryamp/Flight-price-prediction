```python
arrival_time=[]
for w in price['Arrival Time']:
    # print(w)
    w=w.split()
    arrival_time.append(w[0])
    
```

```python
dur_h=[]
dur_m=[]
for d in price['Duration']:
    d=d.split()
    hour=d[0].replace('h','')
    minute=d[1].replace('m','')
    dur_h.append(hour)
    dur_m.append(minute)
price['Duration_hour']=dur_h
price['Duration_minute']=dur_m
    
```

```python
price['Price'].replace({',':''},inplace=True)
```

```python
target=[]
for p in price['Price']:
    p=p.replace(',','')
    p=int(p)
    target.append(p)
```
